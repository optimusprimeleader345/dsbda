```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load Dataset
df = pd.read_csv("movie_dataset.csv")

# Display Dataset
print(df.head())

# Features Used
features = [
    'keywords',
    'cast',
    'genres',
    'director'
]

# Fill Missing Values
for feature in features:
    df[feature] = df[feature].fillna('')

# Combine Features
combined_features = (
    df['keywords'] + ' ' +
    df['cast'] + ' ' +
    df['genres'] + ' ' +
    df['director']
)

# Convert Text into Matrix
cv = CountVectorizer()

count_matrix = cv.fit_transform(combined_features)

# Cosine Similarity
similarity = cosine_similarity(count_matrix)

# Input Movie
movie_name = "Avatar"

# Find Movie Index
movie_index = df[df.title == movie_name].index[0]

# Similar Movies
similar_movies = list(
    enumerate(similarity[movie_index])
)

# Sort Movies
sorted_movies = sorted(
    similar_movies,
    key=lambda x:x[1],
    reverse=True
)

# Display Recommendations
print("Recommended Movies:")

for movie in sorted_movies[1:6]:
    index = movie[0]
    print(df.iloc[index]['title'])
```
